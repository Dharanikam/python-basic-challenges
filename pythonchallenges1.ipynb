```python
# Secret Message Decoder

# Ask the user to input a secret message
message = input("Enter your secret message: ")

# Step 1: Convert entire message to uppercase
message = message.upper()

# Step 2: Replace all spaces with underscores
message = message.replace(" ", "_")

# Step 3: Show only every 3rd character
third_chars = message[::3]

# Step 4: Reverse the message
reversed_msg = message[::-1]

# Step 5: Display results
print("\n--- Secret Message Decoder ---")
print("Original message length:", len(message))
print("Transformed Message:", message)
print("Every 3rd Character:", third_chars)
print("Reversed Message:", reversed_msg)
```
