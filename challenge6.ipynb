```python
print("Welcome to the Shopping Cart System!")
print("Enter item prices. Enter 0 to finish.\n")

subtotal = 0
item_count = 0

while True:
    price = float(input("Enter item price: "))

    # Sentinel value to stop
    if price == 0:
        break

    # Negative price validation
    if price < 0:
        print("Invalid price! Please enter a positive value.")
        continue

    # Add to cart
    subtotal += price
    item_count += 1

# Apply discounts
if subtotal >= 1000:
    discount_rate = 0.20
elif subtotal >= 500:
    discount_rate = 0.10
elif subtotal >= 200:
    discount_rate = 0.05
else:
    discount_rate = 0

discount_amount = subtotal * discount_rate
final_total = subtotal - discount_amount

# Bill summary
print("\n------- BILL SUMMARY -------")
print("Total items purchased:", item_count)
print("Subtotal: ₹", subtotal)
print("Discount applied:", discount_rate * 100, "%")
print("Discount amount: ₹", discount_amount)
print("Final total to pay: ₹", final_total)
print("----------------------------")
```
